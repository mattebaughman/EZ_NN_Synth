```python
from keras import Sequential
from keras.layers import Dense
import pandas as pd
from keras.optimizers import Nadam, SGD
```

```python
from numpy import array
from scipy.stats.mstats import gmean
```

```python
data = pd.read_csv('data.csv',header=[0])
```

```python
data=data.drop('Date',axis=1)
```

```python
for p in data.columns.values:
    data[p]=data[p]-min(data[p])
    data[p]=data[p]/(max(data[p])/2)-1
```

```python
data
```

```python
X=array(data[['amzn_rev','amzn','gcp_rev','alph_rev','azure_rev','msft_rev']])
```

```python
Y=array(data[['aws_rev']])
```

```python
Y
```

```python
model=Sequential()
shapex = X[0].shape
model.add(Dense(shapex[0],input_shape=shapex,activation='selu'))
model.add(Dense(int(gmean([shapex[0],1])),activation='selu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
nadam = Nadam(lr=.00000001)
sgd = SGD(lr=.0001)
model.compile(optimizer=nadam,loss='mae')
```

```python
model.fit(X[4:],Y[4:],epochs=5000)
```
